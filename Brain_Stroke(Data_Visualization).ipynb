```python
#importing libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(font_scale=1.5)
```

```python
#reading dataset
data = pd.read_csv('stroke.csv') 
```

Data Visualization

1. Gender

```python
#Plotting Bar graph for Gender attributes
plt.figure(figsize=(10,12))
dig = sns.countplot(data=data,x='gender')
plt.show()
```

```python
#printing the Male, Female and Other gender ratio for Gender
print('Male:',len(data[data['gender']=='Male']),
      'Female:',len(data[data['gender']=='Female']),
      'Others:',len(data[data['gender']=='Other']))
```

2. Hypertension

```python
#Plotting Bar graph for Hypertension attributes
plt.figure(figsize=(8,10))
dig = sns.countplot(data=data,x='hypertension')
plt.show()
```

```python
#printing the yes and no ratio for hypertension
print('Yes:',len(data[data['hypertension']==1]),
      'No:',len(data[data['hypertension']==0]))
```

3. Work Type

```python
#Plotting Bar graph for Work Type attributes
plt.figure(figsize=(12,12))
sns.countplot(data= data, x='work_type')
plt.show()
```

```python
#printing the attributes ratio for Work Type
print('Private:',len(data[data['work_type']=='Private']),
      'Self-employed:',len(data[data['work_type']=='Self-employed']),
      'Govt_job:',len(data[data['work_type']=='Govt_job']),
      'children:',len(data[data['work_type']=='children']),
      'Never_worked:',len(data[data['work_type']=='Never_worked']))
```

4. Residence Type

```python
#Plotting Bar graph for Residence Type attributes
plt.figure(figsize=(12,12))
sns.countplot(data= data, x='Residence_type')
plt.show()
```

```python
#printing the attributes ratio for Residence Type
print('Urban:',len(data[data['Residence_type']=='Urban']),
      'Rural:',len(data[data['Residence_type']=='Rural']))
```

5. Smoking Status

```python
#Plotting Bar graph for Smoking Status attributes
plt.figure(figsize=(10,12))
sns.countplot(data =data, x = 'smoking_status')
plt.show()
```

```python
#printing the attributes ratio for Smoking Status
print('formerly smoked:',len(data[data['smoking_status']=='formerly smoked']),
      'never smoked:',len(data[data['smoking_status']=='never smoked']),
      'smokes:',len(data[data['smoking_status']=='smokes']),
      'Unknown:',len(data[data['smoking_status']=='Unknown']))
```

6. Stroke

```python
#Plotting Bar graph for Stroke attributes
plt.figure(figsize=(10,12))
sns.countplot(data= data, x='stroke')
plt.show()
```

```python
#printing the attributes ratio for Work Type
print('No stroke:',len(data[data['stroke']==0]),
      'Stroke:',len(data[data['stroke']==1]))
```
